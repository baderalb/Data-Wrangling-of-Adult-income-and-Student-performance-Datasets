```python
# FILL IN - Make copies of the datasets to ensure the raw dataframes 
# are not impacted
student7=student.copy()
adult2=adult_df.copy()
```

### **Quality Issue 1: completness

```python
# FILL IN - Apply the cleaning strategy
adult2.replace('?', np.nan, inplace=True)  # Replace '?' with NaN\
# Optionally drop rows with missing values in workclass and occupation
adult2.dropna(subset=['workclass', 'occupation'], inplace=True)
```

```python
# FILL IN - Validate the cleaning was successful
remaining_missing_workclass = adult2['workclass'].isnull().sum()
remaining_missing_occupation = adult2['occupation'].isnull().sum()

print(f"\nRemaining missing values in 'workclass' after cleaning: {remaining_missing_workclass}")
print(f"Remaining missing values in 'occupation' after cleaning: {remaining_missing_occupation}")
```

Justification: *FILL IN*
 
  Justification: The replacement of '?' with NaN allows for the identification and handling of missing values
 in a standardized way. By dropping rows with missing values in the 'workclass' and occupation' columns,
#we ensure that our dataset only includes entries with complete information for these critical features.
 This improves the quality and reliability of subsequent analyses, as incomplete data can lead to biased
 or inaccurate results.

### **Quality Issue 2: Accuracy

```python
#FILL IN - Apply the cleaning strategy
#Replace unrealistic hours worked >65 with NaN
adult2.loc[adult_df['hours-per-week'] > 65, 'hours-per-week'] = np.nan
#fill with the median
adult2['hours-per-week'].fillna(adult2['hours-per-week'].median(), inplace=True)
```

```python
#FILL IN - Validate the cleaning was successful
adult2['hours-per-week'].describe()#oultieresremovedby looking the max value is 65 
```

Justification: The replacement of unrealistic values (hours worked > 65) with NaN and subsequent
 filling with the median ensures that outliers do not skew the analysis. Working more than 65 hours
 per week is considered unrealistic and can significantly impact the mean and standard deviation.
By filling these NaN values with the median, we retain a central tendency measure that is robust
 to outliers. This process helps in maintaining the integrity of the data and ensures more reliable
 analysis results. The validation step, which checks the descriptive statistics of the 'hours-per-week'
 column, confirms that the maximum value is now within a realistic range.

### **Tidiness Issue 1: each varible is a colum the first rule**

```python
#FILL IN - Apply the cleaning strategy
# Apply the cleaning strategy
# Reshape the dataframe using melt
melted_df = student7.melt(id_vars=['StudentID', 'Age', 'Gender', 'Ethnicity', 'ParentalEducation', 'StudyTimeWeekly', 'Absences', 'Tutoring', 'ParentalSupport', 'GPA', 'GradeClass'], 
                             value_vars=['Extracurricular', 'Sports', 'Music', 'Volunteering'],
                             var_name='Activity', 
                             value_name='Participation')
# Filter out rows where Participation is 0 (i.e., no participation)
melted_df = melted_df[melted_df['Participation'] == 1]
melted_df.head()
melted_df=melted_df.reset_index()
```

```python
#FILL IN - Validate the cleaning was successful
#FILL IN - Validate the cleaning was successful
# Verify the unique values in the ActivityType column
melted_df['Activity'].value_counts()
#now we are sure that we did the melting correct
# Replace the original student5 DataFrame with the melted DataFrame
student7= melted_df

# Now student5 contains the reshaped data

student7.reset_index(drop=True, inplace=True)
student7.drop("index",axis=1,inplace=True)# an index column removed
student7.head()
```


The original DataFrame had separate columns for different types of participation in extracurricular activities (Extracurricular, Sports, Music, and Volunteering), which violated the tidiness principle that each variable should have its own column. By reshaping the DataFrame using the melt function, we transformed these columns into a single 'Activity' column, with each type of activity represented as a value in this column. This approach reduces redundancy and simplifies the analysis of overall participation trends among students. The final validation step confirmed that the melting and filtering processes were correctly applied, as evidenced by the distinct count of each activity type in the 'Activity' column.
